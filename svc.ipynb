```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Lets create synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split

Xtrain,Xtest, ytrain,ytest = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
Xtrain.shape, Xtest.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear')
```

```python
model.fit(Xtrain,ytrain)
```

```python
model.coef_
```

```python
ypred = model.predict(Xtest)
```

```python
ypred
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

print(f"Accuracy Score : \n{accuracy_score(ytest,ypred)*100}")
print("\n===============================================================\n")
print(f"Confusion Matrix : \n{confusion_matrix(ytest,ypred)}")
print("\n===============================================================\n")
print(f"Classification Report : {classification_report(ytest,ypred)}")
```

```python
from sklearn.model_selection import GridSearchCV

parameter = {
            'C': [0.1, 1, 10, 100, 1000],
            'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
            'kernel':['linear']
}
```

```python
grid = GridSearchCV(SVC(),param_grid=parameter, cv=5, refit=True,verbose=3)
```

```python
grid.fit(Xtrain,ytrain)
```

```python
grid.best_params_
```

```python
pridict = grid.predict(Xtest)
```

```python
pridict
```

```python
print(f"Accuracy Score : \n{accuracy_score(pridict,ytest)*100}")
print("\n----------------------------------------------------------------\n")
print(f"Confusion Matrix : \n{confusion_matrix(pridict,ytest)}")
print("\n----------------------------------------------------------------\n")
print(f"Classification Report : \n{classification_report(pridict,ytest)}")
```
